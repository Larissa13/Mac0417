```python
img0 = read_binary_image("images/segmentation-0.png")
img1 = read_binary_image("images/segmentation-1.png")
img2 = read_binary_image("images/segmentation-2.png")
img3 = read_binary_image("images/segmentation-3.png")
img4 = read_binary_image("images/segmentation-4.png")

display_binary_image(img0)
display_binary_image(img1)
display_binary_image(img2)
display_binary_image(img3)
display_binary_image(img4)
```

<p>
    <strong>1.) </strong> Projete um operador morfológico para segmentar objetos com alças. Teste em todas as
    imagens.
</p>

```python
disk = create_structure_element_disk(7)
img0_opening = opening_top_hat(img0, disk)
img0_marker = area_opening(img0_opening, 100)
img0_reconstructed= inf_reconstruction(img0_marker, img0)
display_image(normalize(img0_reconstructed))

```

```python
img1_opening = opening_top_hat(img1, disk)
img1_marker = area_opening(img1_opening, 250)
img1_reconstructed = inf_reconstruction(img1_marker, img1)
display_image(normalize(img1_reconstructed))
```

```python
img2_opening = opening_top_hat(img2, disk)
img2_marker = area_opening(img2_opening, 250)
img2_reconstructed = inf_reconstruction(img2_marker, img2)
display_image(img2_reconstructed)
print(1*img2_reconstructed.min(), 1*img2_reconstructed.max())
```

```python
img3_opening = opening_top_hat(img3, disk)
img3_marker = area_opening(img3_opening, 1000)
img3_reconstructed = inf_reconstruction(img3_marker, img3)
display_image(normalize(img3_reconstructed))
```

```python
img4_opening = opening_top_hat(img4, disk)
img4_marker = area_opening(img4_opening, 500)
img4_reconstructed = inf_reconstruction(img4_marker, img4)
display_image(normalize(img4_reconstructed))
```

<p>
    <strong> 2.)</strong> Projete um operador morfológico para segmentar besouros. Teste em todas as imagens.
</p>

```python
box = create_structure_element_box()
img0_opening2 = opening_top_hat(img0, box)
img0_marker2 = area_opening(img0_opening2, 50)
img0_reconstructed2 = inf_reconstruction(img0_marker2, img0)
display_image(img0_reconstructed2)
```

```python
img1_opening2 = opening_top_hat(img1, box)
img1_marker2 = area_opening(img1_opening2, 100)
img1_reconstructed2 = inf_reconstruction(img1_marker2, img1)
display_image(img1_reconstructed2)
```

```python
box2 = create_structure_element_box(5)
img2_opening2 = opening_top_hat(img2, box2)
img2_marker2 = area_opening(img2_opening2, 200)
img2_reconstructed2 = inf_reconstruction(img2_marker2, img2)
display_image(normalize(img2_reconstructed2))
```

```python
img3_opening2 = opening_top_hat(img3, box2)
img3_marker2 = area_opening(img3_opening2, 300)
img3_reconstructed2 = inf_reconstruction(img3_marker2, img3)
display_image(normalize(img3_reconstructed2))
```

```python
img4_opening2 = opening_top_hat(img4, box2)
img4_ao1 = area_opening(img4_opening2, 400)
img4_ao2 = area_opening(img4_opening2, 500)
img4_marker2 = img4_ao1 ^ img4_ao2 #intersecção
img4_reconstructed2 = inf_reconstruction(img4_marker2, img4)
display_image(normalize(img4_reconstructed2))
```
